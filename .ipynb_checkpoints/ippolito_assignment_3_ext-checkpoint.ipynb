```python
# Load libraries
import networkx as net
import requests
import os.path
import matplotlib.pyplot as plt
import numpy as np
```

Load Paris metro data taken from https://github.com/BTajini/Paris-Metro-Project and includes a graph of the Paris metro, along with weights along each edge.

We'll read the data from disk if available; otherwise we'll get it from the git repo. Then we'll parse the data, add each node and edge to the networkx graph, and display the graph.

```python
# Read data
if os.path.exists('metro_paris.txt'):
    print('reading data from disk')
    fh = open('metro_paris.txt', 'r', encoding='utf-8')
    r = fh.read()
    fh.close()
else:
    url = 'https://raw.githubusercontent.com/BTajini/Paris-Metro-Project/master/Data/metro_paris.txt'
    print('fetching data from ' + url)
    r = requests.get(url)
    r = r.text
    fh = open('metro_paris.txt', 'w', encoding='utf-8')
    fh.write(r)
    fh.close()

# Split into nodes and edges
data = r.split('[Edges]\n')
nodes = dict([l.split(' ', maxsplit=1) for l in data[0].split('\n')[1:-1]])    # don't use first line "[Vertices]" or last (blank) line
edges = [l.split(' ') for l in data[1].split('\n')]
for e in edges:
    e[0] = e[0].rjust(4, '0')
    e[1] = e[1].rjust(4, '0')
    e[2] = float(e[2])
print('Data includes', len(nodes), 'nodes and', len(edges), 'edges')

# Print first six samples of each
print("Data sample:")
print("\tnodes:",[(k, nodes[k]) for k in nodes.keys()][:6])
print("\tedges:",edges[:6])
```

```python
# Create new graph
g = net.Graph()

# Add edges
for e in edges:
    #g.add_edge(nodes[e[0]], nodes[e[1]], weight=e[2])
    g.add_edge(e[0], e[1], weight=e[2])

# Draw the graph
net.draw(g, with_labels=False, node_size=10, font_size=8, alpha=0.5, width=1)
```

#### Graph metrics
Find the graph diameter, degree min and max, and closeness centrality.


```python
# Graph diameter
print("diameter=" + str(net.diameter(g)))
print()

# Degree min and max
deg = dict(net.degree(g))
print("degree min=" + str(min(deg.values())))
print("degree max=" + str(max(deg.values())))
print()

# Closeness centrality
c = net.closeness_centrality(g)
cs = sorted(c.items(), key=lambda item: item[1], reverse=True)
print("closeness centrality (top 10):")
for i in range(0, 10):
    print('\t', nodes[cs[i][0]], cs[i][1])
print("Note: some node names are repeated because some stations have different stops depending on the metro line\n" + \
    "\tand are listed as separate nodes in the data.")
print()
```

```python
# Dijsktra algorithm to find shortest path from one node to all the rest
def shortest_path(node0, nodes, edges):

    # Build dict keyd on nodes, with values containing adjacent nodes, cost, and path
    dcost = {}
    for k in nodes.keys():
        dcost[k] = {}
        dcost[k]['cost'] = np.Inf              # initially set cost to infinity
        dcost[k]['path'] = []                  # initially set empty path
        dcost[k]['adj'] = {}                   # build list of adjacent nodes based on edges list
        for e in edges:
            if k == e[0]:
                dcost[k]['adj'][e[1]] = e[2]
            elif k == e[1]:
                dcost[k]['adj'][e[0]] = e[2]
    
    # Create empty list of visited verticies
    visited = []

    # Set starting node
    thisnode = node0       # which node we're working on
    thiscost = 0           # the cost to get to this node

    # Walk the graph
    while True:

        # Set the cost to be the current cost
        dcost[thisnode]['cost'] = thiscost
        print(thisnode)
        
        # Set the node as visited
        visited.append(thisnode)

        # Walk adjacencies
        for k in dcost[node0]['adj'].keys():

            # Update the cost list if the cost to this adjacency is lower than the existing cost
            if (thiscost + dcost[node0]['adj'][k]) < dcost[k]['cost']:
                dcost[k]['cost'] = thiscost + dcost[node0]['adj'][k]

        #print(dcost)

        # Choose the next node with the lowest cost
        thisnode = ''
        mincost = np.Inf
        for k in dcost.keys():
            if k not in visited:
                
                # If the cost for this node is less than the min, set it to be the next node
                if dcost[k]['cost'] < mincost:
                    mincost = dcost[k]['cost']
                    thisnode = k

        # Escape clause: if the nodes have already been visited or the costs are all higher, exit the while loop
        print("\tnext node:", thisnode)
        if thisnode == '':
            break
```
